```python
import psycopg2
import pandas as pd

# String de conexão
conn_string = "postgresql://externo.leitura:WdN#R10gX4@172.31.3.62:5432/fiscalize"

# Conectar ao banco
conn = psycopg2.connect(conn_string)
cursor = conn.cursor()

# Executar query com JOIN para filtrar pela situacao da demanda
query = """
SELECT fd.* FROM public.fiscaldemanda fd
JOIN public.demanda d ON fd.demanda_id = d.id
WHERE fd.usuario_id = 12201 and fd.ativo = true and d.ativo = true and d.situacao IN (2, 11, 12);
"""
cursor.execute(query)

# Pegar os nomes das colunas
colnames = [desc[0] for desc in cursor.description]

# Pegar os resultados
rows = cursor.fetchall()

# Fechar conexão
cursor.close()
conn.close()

# Criar DataFrame
df = pd.DataFrame(rows, columns=colnames)

# Configurar opções de exibição do pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)

# Exibir o DataFrame com melhor formatação
df
```
