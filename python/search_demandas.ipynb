```python
# Query: Demandas do usuário 12201
conn = psycopg2.connect(CONN_STRING)
cursor = conn.cursor()

query = """
WITH operacoes_validas AS (
    SELECT
        fo.operacao_id
    FROM
        fiscaloperacao fo
    WHERE
        fo.ativo
        AND fo.datainicio < now()
        AND fo.datafim > now()
        AND fo.usuario_id = 12201
)
SELECT d.protocolo,
    d.operacao_id
FROM demanda d
    LEFT JOIN operacao o ON o.id = d.operacao_id
WHERE
    d.ativo
    AND d.situacao = 11
    AND d.operacao_id IN (
        SELECT operacao_id FROM operacoes_validas
    );
"""

# query = """

#  SELECT
#         fo.operacao_id
#     FROM
#         fiscaloperacao fo
#     WHERE
#         fo.ativo
#         AND fo.datainicio < now()
#         AND fo.datafim > now()
#  """

cursor.execute(query)

colnames = [desc[0] for desc in cursor.description]
rows = cursor.fetchall()

cursor.close()
conn.close()

df = pd.DataFrame(rows, columns=colnames)
df
```
